# Theory on the Gradient Descent algorithm

```python
# Function to be optimized
import numpy as np
import matplotlib.pyplot as plt

# Function to be optimized
def f_sphere(x, y):
    return x**2 + y**2

# Gradient of the function
def grad_f_sphere(x, y):
    grad_x = 2 * x
    grad_y = 2 * y
    return np.array([grad_x, grad_y])
```

```python
# Run gradient descent algorithm
def gradient_descent(learning_rate=0.1, iterations=100):
    x, y = 10, 10  # Initial point
    history = [(x, y)]
    for _ in range(iterations):
        grad = grad_f_sphere(x, y)
        x -= learning_rate * grad[0]
        y -= learning_rate * grad[1]
        history.append((x, y))
    return np.array(history)

# Run the gradient descent algorithm
history = gradient_descent()
```

```python
# Plot the results
# Define a meshgrid for plotting the function
x_range = np.arange(-15, 15, 0.1)
y_range = np.arange(-15, 15, 0.1)
X, Y = np.meshgrid(x_range, y_range)
Z = f_sphere(X, Y)

# Plot the surface and the gradient descent path
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='coolwarm', alpha=0.5)

# Plot the path of the gradient descent
ax.plot(history[:, 0], history[:, 1], f_sphere(history[:, 0], history[:, 1]), 'r-o')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
plt.show()
```

```python
# Plot the path of the gradient descent
ax.plot(history[:, 0], history[:, 1], f_sphere(history[:, 0], history[:, 1]), 'r-o')

# Annotate the starting point
ax.text(history[0, 0], history[0, 1], f_sphere(history[0, 0], history[0, 1]), 'Start', color='green')

# Annotate the final point
ax.text(history[-1, 0], history[-1, 1], f_sphere(history[-1, 0], history[-1, 1]), 'End', color='red')

```

```python
# Plot the path of the gradient descent with a gradient of colors
points = ax.scatter(history[:, 0], history[:, 1], f_sphere(history[:, 0], history[:, 1]), c=np.linspace(0, 1, len(history)), cmap='jet')
fig.colorbar(points, ax=ax, label='Iteration')
```

```python
# Plot the surface of the function
# The cmap 'coolwarm' is chosen for its clear distinction between high and low values, making the surface easier to interpret
ax.plot_surface(X, Y, Z, cmap='coolwarm', alpha=0.5)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define la función y su gradiente
def f_sphere(x, y):
    return x**2 + y**2

def grad_f_sphere(x, y):
    grad_x = 2 * x
    grad_y = 2 * y
    return np.array([grad_x, grad_y])

# Implementa el algoritmo de descenso por gradiente
def gradient_descent(learning_rate=0.1, iterations=100):
    x, y = 10, 10  # Punto inicial
    history = [(x, y)]
    for _ in range(iterations):
        grad = grad_f_sphere(x, y)
        x -= learning_rate * grad[0]
        y -= learning_rate * grad[1]
        history.append((x, y))
    return np.array(history)

# Ejecuta el algoritmo de descenso por gradiente
history = gradient_descent()

# Define una malla para graficar la función
x_range = np.arange(-15, 15, 0.1)
y_range = np.arange(-15, 15, 0.1)
X, Y = np.meshgrid(x_range, y_range)
Z = f_sphere(X, Y)

# Crea la figura y los ejes
fig = plt.figure()

# Gráfico 3D
ax = fig.add_subplot(121, projection='3d')
ax.plot_surface(X, Y, Z, cmap='coolwarm', alpha=0.5)

# Traza el camino del descenso por gradiente
ax.plot(history[:, 0], history[:, 1], f_sphere(history[:, 0], history[:, 1]), 'r-o')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')

# Gráfico de contorno en 2D
ax2 = fig.add_subplot(122)
contour = ax2.contourf(X, Y, Z, 50, cmap='viridis')
ax2.plot(history[:, 0], history[:, 1], 'ro-', markersize=5)
ax2.set_xlabel('X axis')
ax2.set_ylabel('Y axis')
ax2.set_title('Contorno de la Función 2D con Caminos de Optimización')
fig.colorbar(contour, ax=ax2)

# Muestra los gráficos
plt.show()
```
